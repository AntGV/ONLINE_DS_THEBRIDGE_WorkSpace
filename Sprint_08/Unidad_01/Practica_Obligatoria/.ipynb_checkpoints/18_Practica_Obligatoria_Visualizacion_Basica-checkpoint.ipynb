### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_popmax = df_pop.nlargest(columns="population_total", n=5)[["city","population_total"]]
df_popmax
```

```python
# Gráfico 1

#plt.figure(figsize = (6,7))
#plt.bar(cities,population,**kwargs)
#plt.title("Ciudades más pobladas de California")
#plt.yticks(ticks = [0,1000000,2000000,3000000,4000000,5000000], labels = [0,1000,2000,3000,4000,5000])
#plt.xlabel("Ciudades")
#plt.ylabel("Habitantes(miles)")

cities = df_popmax["city"]
population = df_popmax["population_total"]
plt.rcParams["font.size"] = 8
fig1,ax1 = plt.subplots()
kwargs1 = {
    "color":["blue","gray","red","olive","brown"],
    "alpha":0.45
}

#ax1.figure.set_size_inches(6,7)
ax1.bar(cities,population, **kwargs1)
ax1.set_title("Ciudades más pobladas de California")
ax1.set_xlabel("Ciudades")
ax1.set_ylabel("Habitantes(miles)");
ax1.set_yticks(ticks = [0,1000000,2000000,3000000,4000000,5000000], labels = [0,1000,2000,3000,4000,5000])

for i in range(len(population.values)):
    ax1.text(i,(population.values[i] + 100000),population.values[i], ha = "center")
    
```

```python
# Gráfico 2

#plt.hist(df_pop["area_total_km2"], **kwargs2);
#plt.xlabel("Superficie ($km^2$)", family =  "Bradley Hand itc", size = 18)
#plt.ylabel("Número de ciudades", family =  "Bradley Hand itc", size = 18)
#plt.legend();

fig2,ax2 = plt.subplots()
kwargs2 = {
    "bins":50,
    "color":"steelblue",
    "alpha":0.7    
}

ax2.hist(df_pop["area_total_km2"], **kwargs2)
ax2.set_xlabel("Superficie ($km^2$)", family =  "Bradley Hand itc", size = 18)
ax2.set_ylabel("Número de ciudades", family =  "Bradley Hand itc", size = 18)
ax2.legend(["Hist Area"], loc = "upper right", fontsize = 8);
```

```python
# Gráfico 3

plt.rcParams["font.size"] = 8
fig3,ax3 = plt.subplots()

ax3.scatter(df_pop["area_water_km2"], df_pop["population_total"], color = "green", alpha = 0.5)
ax3.set_title("Población por kilómetro cuadrado de agua")
ax3.set_xlabel("Área de agua en la ciudad (km$^2$)")
ax3.set_ylabel("Población (miles habitantes)")
ax3.set_yticks(ticks = [0,500000,1000000,1500000,2000000,2500000,3000000,3500000,4000000], labels = [0,500,1000,1500,2000,2500,3000,3500,4000]);

props_arrow = {
    "arrowstyle":"->",
    "linewidth":0.5
}

x = df_pop.nlargest(columns="population_total", n=1)["area_water_km2"]
x1 = x.iloc[0]
y1 = df_pop["population_total"].max()
x2 = df_pop["area_water_km2"].max()
y = df_pop.nlargest(columns="area_water_km2", n=1)["population_total"]
y2 = y.iloc[0]
ax3.annotate("Outliers", xy = (x1,y1), xytext = (60,3500000), arrowprops = props_arrow)
ax3.annotate("         ", xy = (x2,y2), xytext = (60,3500000), arrowprops = props_arrow);
```

```python
# Gráfico 4

plt.rcParams["font.size"] = 8
fig4,ax4 = plt.subplots()

kwargs3 = {
    "color":"black",
    "linewidth":0.5
}

df_popalt = df_pop["elevation_m"].dropna()
data = df_popalt.values
ax4.boxplot(data, vert = False, patch_artist = True, boxprops=dict(facecolor="red", **kwargs3), medianprops=dict(**kwargs3), whiskerprops=dict(**kwargs3))
ax4.set_xlabel("Altitud (m)");
```

```python
# Juntar los cuatro gráficos

fig,ax = plt.subplots(2,2, figsize =(13,13))

ax[0,0]
ax[0,0].bar(cities,population, **kwargs1)
ax[0,0].set_title("Ciudades más pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes(miles)");
ax[0,0].set_yticks(ticks = [0,1000000,2000000,3000000,4000000,5000000], labels = [0,1000,2000,3000,4000,5000])
for i in range(len(population.values)):
    ax[0,0].text(i,(population.values[i] + 100000),population.values[i], ha = "center")

ax[0,1]
ax[0,1].hist(df_pop["area_total_km2"], **kwargs2)
ax[0,1].set_xlabel("Superficie ($km^2$)", family =  "Bradley Hand itc", size = 18)
ax[0,1].set_ylabel("Número de ciudades", family =  "Bradley Hand itc", size = 18)
ax[0,1].legend(["Hist Area"], loc = "upper right", fontsize = 8)

ax[1,0]
ax[1,0].scatter(df_pop["area_water_km2"], df_pop["population_total"], color = "green", alpha = 0.5)
ax[1,0].set_title("Población por kilómetro cuadrado de agua")
ax[1,0].set_xlabel("Área de agua en la ciudad (km$^2$)")
ax[1,0].set_ylabel("Población (miles habitantes)")
ax[1,0].set_yticks(ticks = [0,500000,1000000,1500000,2000000,2500000,3000000,3500000,4000000], labels = [0,500,1000,1500,2000,2500,3000,3500,4000])
ax[1,0].annotate("Outliers", xy = (x1,y1), xytext = (60,3500000), arrowprops = props_arrow)
ax[1,0].annotate("         ", xy = (x2,y2), xytext = (60,3500000), arrowprops = props_arrow)

ax[1,1]
ax[1,1].boxplot(data, vert = False, patch_artist = True, boxprops=dict(facecolor="red", **kwargs3), medianprops=dict(**kwargs3), whiskerprops=dict(**kwargs3))
ax[1,1].set_xlabel("Altitud (m)");
```
