```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as shc
import seaborn as sns

from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans
from sklearn.neighbors import KNeighborsClassifier
```

```python
def plot_dbscan(dbscan, X, size, show_xlabels=True, show_ylabels=True):
    core_mask = np.zeros_like(dbscan.labels_, dtype=bool)
    core_mask[dbscan.core_sample_indices_] = True
    anomalies_mask = dbscan.labels_ == -1
    non_core_mask = ~(core_mask | anomalies_mask)

    cores = dbscan.components_
    anomalies = X[anomalies_mask]
    non_cores = X[non_core_mask]
    
    plt.scatter(cores[:, 0], cores[:, 1],
                c=dbscan.labels_[core_mask], marker='o', s=size, cmap="Paired")
    plt.scatter(cores[:, 0], cores[:, 1], marker='*', s=50, c=dbscan.labels_[core_mask])
    plt.scatter(anomalies.iloc[:, 0], anomalies.iloc[:, 1],
                c="r", marker="x", s=20)
    plt.scatter(non_cores.iloc[:, 0], non_cores.iloc[:, 1], c=dbscan.labels_[non_core_mask], marker=".")
    if show_xlabels:
        plt.xlabel(X.columns[0], fontsize=14)
    else:
        plt.tick_params(labelbottom=False)
    if show_ylabels:
        plt.ylabel(X.columns[1], fontsize=14, rotation=0)
    else:
        plt.tick_params(labelleft=False)
    plt.title("eps={:.2f}, min_samples={}".format(dbscan.eps, dbscan.min_samples), fontsize=14)
```

```python
def cluster_noise(dbscan):
    labels = dbscan.labels_
    
    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)

    print("Número estimado de clusters: %d" % n_clusters_)
    print("Número estimado de puntos sin agrupar: %d" % n_noise_)
```

```python
def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1)
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)
```

```python
def plot_dendrogram(df):
    plt.figure(figsize = (10, 7))
    plt.title(f"Dendrograma de '{df.columns[0]}' y '{df.columns[1]}'")
    
    selected_data = df
    clusters = shc.linkage(selected_data, method = 'ward', metric = "euclidean")
    shc.dendrogram(Z = clusters)
    plt.show()
```
